```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
web_scrap = 'https://ascopubs.org/jco/special/meeting-abstracts?expanded=tvolume-suppl.d2020.y2021&expanded=tvolume-suppl.d2020.y2021.v39'
Web_asco = web_scrap + '/events/'
```

```python
html_file = requests.get(Web_asco)
html_file
```

```python
soup_asco = BeautifulSoup(html_file.text,'lxml')
print(soup_asco.prettify())
```

```python
for i in soup_asco.find_all('div', class_='js_issue row'):
    for j in i.find_all('a'):
        print('https://ascopubs.org' + j.get('href'))
```

```python
asco_meeting_abstract_2022 = []

for i in soup_asco.find_all('div', class_='issueGroup'):
    for j in i.find_all('div', class_='slider opened', id ='tvolume-suppl.d2020.y2022.v40'):
        for k in j.find_all('div', class_='js_issue row'):
            for s in k.find_all('a'):
                asco_meeting_abstract_2022.append('https://ascopubs.org' + s.get('href'))

asco_meeting_abstract_ab = []                
                
for i in range(5):
    asco_meeting_abstract_ab.append(asco_meeting_abstract_2022[i])
    print(asco_meeting_abstract_2022[i])
```

```python
links_data_abstract = []
for i in range(len(asco_meeting_abstract_ab)):
    internal_link_1 = []
    a = asco_meeting_abstract_ab[i]
    
    b = requests.get(a)
    c = BeautifulSoup(b.text, 'lxml')
    
    for p in c.find_all('div', class_='tocDeliverFormatsLinks'):
        for q in p.find_all('a', class_='ref nowrap abs'):
            internal_link_1.append('https://ascopubs.org' + q.get('href'))
            
    links_data_abstract.append(internal_link_1)
links_data_abstract
```

```python
for i in range(len(links_data_abstract)):
    print(len(links_data_abstract[i]))
```

```python
a = []

for i in range(1):
    abstract_data = []
    for j in range(len(links_data_abstract[i])):
        name_authors = []
        
        p = links_data_abstract[i][j]
        abstract_data.append(p)
        
        q = requests.get(p)
        r = BeautifulSoup(q.text, 'lxml')
        
        for p in r.find_all('div', class_='publicationContentTitle'):
            for q in p.find_all('h1', class_='chaptertitle'):
                abstract_data.append(q.text)
                
        for x in r.find_all('span', class_='contribDegrees'):
            for y in x.find_all('a', class_='entryAuthor'):
                name_authors.append(y.text)
        del name_authors[1:len(name_authors):2]
        abstract_data.append(name_authors)
        
        for m in r.find_all('div', class_='abstractSection abstractInFull'):
            abstract_data.append(m.text)
    a.append(abstract_data)    
a  
```
